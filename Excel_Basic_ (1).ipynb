1.  Q1. Explain the difference between Relative, Absolute and Mixed Cell Referencing
Answer:
Relative Cell Referencing means the cell reference changes automatically when the formula is copied to another cell.
Example: =A1+B1
Absolute Cell Referencing means the cell reference remains fixed even after copying the formula.
Example: =$A$1+$B$1
Mixed Cell Referencing means either row or column is fixed.
Example: =$A1 or =A$1
